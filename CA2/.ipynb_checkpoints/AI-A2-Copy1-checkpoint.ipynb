```python
import librosa
import numpy as np
import os
import soundfile as sf
import matplotlib.pyplot as plt
import noisereduce as nr
```

# Step One:

## Preprocessing:

```python
def preprocess_audio(file_path, output_path, target_sampling_rate=16000)::
    # Load the audio file
    audio, sampling_rate = librosa.load(file_path, sr=target_sampling_rate)

    # Noise reduction
    reduced_noise_audio = nr.reduce_noise(y=audio, sr=sampling_rate)

    # Silence removal
    non_silent_audio, _ = librosa.effects.trim(reduced_noise_audio, top_db=20) # top_db: threshold for silence in dB

    # Normalize the audio to a standard volume level
    normalized_audio = librosa.util.normalize(non_silent_audio)

    # Save the processed audio file
    sf.write(output_path, normalized_audio, target_sampling_rate)
```

```python
def preprocess_all_audio(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for file_name in os.listdir(input_folder):
        if file_name.endswith('.wav'):
            file_path = os.path.join(input_folder, file_name)
            output_path = os.path.join(output_folder, file_name)
            preprocess_audio(file_path, output_path)
            print(f"Processed {file_name}")
```

```python

input_folder = 'recordings'
output_folder = 'preprocessed_recordings'
preprocess_all_audio(input_folder, output_folder)
```

## Feature Extraction

### Extract MFCC

```python
def extract_mfcc(file_path, n_mfcc=13):
    audio, sample_rate = librosa.load(file_path, sr=None)
    mfcc_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=n_mfcc)
    return mfcc_features.T
```

```python
def extract_features_for_all_files(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for file_name in os.listdir(input_folder):
        if file_name.endswith('.wav'):
            file_path = os.path.join(input_folder, file_name)
            mfcc_features = extract_mfcc(file_path)
            
            # Saving MFCC features to a file
            output_file_path = os.path.join(output_folder, file_name.replace('.wav', '.npy'))
            np.save(output_file_path, mfcc_features)
            print(f"MFCC features extracted and saved for {file_name}")
```

```python
input_folder = 'preprocessed_recordings'
output_folder = 'mfcc_features'
extract_features_for_all_files(input_folder, output_folder)
```

### Heat Map:

```python
def plot_mfcc_heatmaps(mfcc_dict):
    for file_name, mfcc in mfcc_dict.items():
        plt.figure(figsize=(25, 10))
        plt.title(f'MFCC Heatmap for {file_name}')
        librosa.display.specshow(mfcc, x_axis='time')
        plt.colorbar(format='%+2.00f dB')
        plt.tight_layout()
        plt.show()
```

```python
mfcc_folder = 'mfcc_features'
index_of_interest = 0 
files_with_index_0 = [f for f in os.listdir(mfcc_folder) if f.endswith('.npy') and f.split('_')[-1].replace('.npy', '') == str(index_of_interest)]
mfcc_features = {file: np.load(os.path.join(mfcc_folder, file)) for file in files_with_index_0}
plot_mfcc_heatmaps(mfcc_features)
```
